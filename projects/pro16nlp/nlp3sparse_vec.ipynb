```python
# 희소벡터 표현 방식
# CountVector - 단어 등장 횟수 기반 - 정수로 표현
# TfidVectorizer : 문서에서 자주 나타는 단어 가중치를 높게 주되, 모든 문서서 자주 등장하는 단어에는 패널티 부여 - 실수 표현
# CountVector : 단어가 몇 번 나왔는가?
# TfidVectorizer : 그 단어가 문서에서 얼마나 중요한가?
# Word2Vec : 그 단어 어떤 의미적 위치에 있는가?

# CountVectorizer와 TfidVectorizer는 '문장 안에 어떤 단어 들어 있는가?'를 숫자 바꾸는 방식
# Word2Vec는 '단어가 어떤 의미적 위치 있는가'를 숫자로 바꾸는 방식
# 키워가 정확히 맞아야하는 검색에는 CountVector, TfidVectorizer가 효과적이고,
# 의미 비슷한 단어까지 찾고 싶다면 Word2Vec이 적합하다.
```

```python
print('CountVectorizer 연습')
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

content = ['How to format my hard disk', 'Hard disk format format problems']
count_vec = CountVectorizer(analyzer='word', min_df=1)
# print(count_vec)
tran = count_vec.fit_transform(content)
print(tran)
print(count_vec.get_feature_names_out())
print(tran.toarray())
```

```python
print('TfidfVectorizer 연습')
print(content)
tfidf_vec = TfidfVectorizer(analyzer='word', min_df=1)
tran2 = tfidf_vec.fit_transform(content)
print(tran2)
print(tfidf_vec.get_feature_names_out())
print(tran2.toarray())
```
