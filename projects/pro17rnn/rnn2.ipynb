```python
# LSTMмңјлЎң лӢӨмқҢ мҲ«мһҗ мҳҲмёЎ
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, Dense, Input

x = np.array([[1, 2, 3], [2, 3, 4], [3, 4, 5], [4, 5, 6], [5, 6, 7], [6, 7, 8], [7, 8, 9], [8, 9, 10]])
y = np.array([4, 5, 6, 7, 8, 9, 10, 11])

x = x.reshape((8, 3, 1))   # мһ…л Ҙнҳ•нғң(samples, time_steps, features)

model = Sequential()
model.add(Input(shape=(3, 1)))   # мӢңкі„м—ҙ(time_steps) 3к°ң, features=1
model.add(LSTM(32, activation='tanh'))
# model.add(LSTM(10, activation='tanh', return_sequences=True)) # н‘ңмӨҖ
# model.add(GRU(10, activation='tanh', return_sequences=True)) # кі„мӮ°лҹү м Ғкі  мҶҚлҸ„лҠ” л№ лҰ„
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='linear'))
model.summary()
```

```python
model.compile(loss='mse', optimizer='adam')

from tensorflow.keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='loss', patience=3, mode='auto')
model.fit(x, y, epochs=1000, batch_size=4, callbacks=[es])
```

```python
print('мҳҲмёЎк°’ :', model.predict(x).ravel())
print('мӢӨм ңк°’ :', y.ravel())

# мғҲлЎңмҡҙ к°’мңјлЎң мҳҲмёЎ
x_input = np.array([0, 1, 2])
x_input = x_input.reshape((1, 3, 1))

new_pred = model.predict(x_input)
print(new_pred.ravel())
```
